```python
# TASK 02 â€“ Titanic Dataset EDA

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("train.csv")

# Overview
print(df.shape)
print(df.info())
print(df.describe())

# Missing values
print(df.isnull().sum())

# Fill missing values
df['Age'].fillna(df['Age'].median(), inplace=True)
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)

# Visualization 1: Survival count
sns.countplot(x='Survived', data=df)
plt.title("Survival Count")
plt.show()

# Visualization 2: Survival by Gender
sns.countplot(x='Sex', hue='Survived', data=df)
plt.title("Survival by Gender")
plt.show()

# Visualization 3: Age Distribution by Survival
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='Age', hue='Survived', kde=True, bins=30, multiple='stack')
plt.title("Age Distribution by Survival")
plt.xlabel("Age")
plt.ylabel("Count")
plt.legend(title="Survived", labels=["0", "1"])
plt.show()
```
