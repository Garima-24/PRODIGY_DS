```python
import pandas as pd

# Load the dataset (skip the first 4 metadata rows)
df = pd.read_csv("API_SP.POP.TOTL_DS2_en_csv_v2_38144.csv", skiprows=4)

# Show the first few rows
df.head()
```

```python
# Keep only the Country and 2022 population columns
df_2022 = df[["Country Name", "2022"]].dropna()
df_2022.columns = ["Country", "Population"]

# Convert population to integers
df_2022["Population"] = df_2022["Population"].astype(int)

# Preview the cleaned data
df_2022.head()
```

```python
import matplotlib.pyplot as plt

# Get the top 10 most populous countries
top10 = df_2022.sort_values(by="Population", ascending=False).head(10)

# Plot bar chart
plt.figure(figsize=(10, 6))
plt.bar(top10["Country"], top10["Population"] / 1e6)
plt.title("Top 10 Most Populous Countries (2022)")
plt.ylabel("Population (in millions)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
from google.colab import files
uploaded = files.upload()

# TASK 01 – Distribution Visualization

import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("API_SP.POP.TOTL_DS2_en_csv_v2_38144.csv", skiprows=4)

# Select recent year and top 10 countries by population
df_latest = df[['Country Name', '2022']]
df_latest = df_latest.dropna()
df_latest = df_latest.sort_values('2022', ascending=False).head(10)

# Plot Bar Chart
plt.figure(figsize=(12, 6))
plt.bar(df_latest['Country Name'], df_latest['2022'] / 1e6, color='teal')
plt.title("Top 10 Most Populated Countries in 2022")
plt.ylabel("Population (in Millions)")
plt.xlabel("Country")
plt.xticks(rotation=45)
plt.tight_layout()
plt.grid(axis='y')
plt.show()
```
